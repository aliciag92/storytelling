**1. Prepare to Create**
- 5 Minutes
- Use NO data: "When you start with the data set in mind, it limits how you think." - Jeff Heer, associate professor of computer science.
- Walk away with a sketch of the presentation/visualization(s).

**2. Talk and Listen**
- 15 Minutes
- Talk to a colleague or friend in order for you to explain basic information, organize thoughts, provide context, and listen to yourself and your friend/colleague for keywords, confusions, etc.

**3. Sketch**
- 20 Minutes
- Match keywords to approaches (Relationship; Comparison; Distribution; Composition; Maps, Networks, Logic).
- Answer questions about the types of information.
- Start sketching: Quick and messy, don't worry about scale, don't add details.

**4. Prototype**
- 20 Minutes
- Add in some data, just enough needed to prototype (like axis scale makes sense to the data, e.g.).
- Create a non-digital prototype/drawing only as detailed as is needed to take it and convert it to a digital version.

# Prepare to Create
1. Identify the **story framework** that fits this project, and use the list of items that make up that framework option (A or B) as a guide for what to answer, capture thoughts or solutions to.
2. Identify/know your **audience** by addressing the questions suggested.
3. Know the **setting** by addressing the questions suggested.
4. **Brainstorm** using sticky notes, refine your ideas, and then generate a visual, non-digital outline of the content.
5. Identify the **type of visualization** by sketching the "Visualization Quadrants" (Conceptual, Data-driven, Declarative, Exploratory) and, for each visualization you are creating, identify where that viz sits.

### Story Framework
>Things to answer/do and document:
>1. Identify the story framework that fits this project.
>2. What is the problem you are trying to solve?
>3. What is my goal for this visualization & presentation?
>4. What type of visualization could demonstrate the problem/issue?
>6. What recommendation are you going to make?
>7. What do I want them to do about it?
>8. What type of visualization could best influence your audience to take action on your recommendation?
>9. How could you visualize what would happen after implementing the recommendation?

- Option A:
    1. Define the issue.
    2. Demonstrate the issue: think about what type of visualization would demonstrate the issue.
    3. Show what happens after implementing a change, a study, etc. and compare.
    4. Make recommendations.
- Option B:
    1. Plot
    2. Rising Action
    3. Climax
    4. Falling Action
    5. Ending

### Audience
>Things to answer/do and document:
>1. Who is your audience?
>2. Have they ever seen this kind of chart before?
>3. What do they already know?
>4. Who else might see it and how?

### Setting
>Things to answer/do and document:
>1. What setting will it be used in?
>2. How much time do I have?
>3. How will it be displayed?
>4. How big is the room?

### Brainstorm
>Things to answer/do and document:
>1. Braindump: Get a stack of sticky notes and write down potential pieces of content. This is simply a process of getting all of the ideas out of your head and onto paper, without any concern for whether or not they make it into the final presentation.
>2. Refine ideas: Group related topics, consider what overall structure will help pull them together in a way that will hopefully make sense to someone else, and discard ideas that don't work.
>3. Generate a visual, non-digital outline of the content. This should demonstrate what ideas/topics will sit where, in what form (text, metric, chart, ...), etc.

# Sketch
>Things to answer/do and document:
>1. What type of information are you communicating?
>2. Map keywords to approaches/types of visualization to identify the chart type(s) that best convey your message.
>3. Sketch chart(s) quickly, without scale, without details.

Type of information:

- Quantitative or Qualitative data
- How is it exposed: numerically, textually, geospatially, relationally?
- Can the data be measured or counted (discrete data (counted), continuous data (measured))?
- Does order matter (ordered data, unordered data)?
- Are you visualizing observations or statistics about the observations (observations, statistics)?

Keywords to approaches: Andy Kriebel's Visual Vocabulary
   - **Deviation**: Emphasise variations (+/-) from a fixed reference point. Typically the reference point is zero but it can also be a target or a long-term average. Can also be used to show sentiment (positive/neutral/negative).
    
   - **Distribution**: Show values in a dataset and how often they occur. The shape (or 'skew') of a distribution can be a memorable way of highlighting the lack of uniformity or equality in the data.
    
   - **Magnitude**: Show size comparisons. These can be relative (just being able to see larger/bigger) or absolute (need to see fine differences). Usually these show a 'counted' number (for example, barrels, dollars or people) rather than a calculated rate or per cent.
    
   - **Correlation**: Show the relationship between two or more variables. Be mindful that, unless you tell them otherwise, many readers will assume the relationships you show them to be casual (i.e., one causes the other).
   - **Change over Time**: Give emphasis to changing trends. These can be short (intra-day) movements or extended series traversing decades or centuries: Chossing the correct time period is important to provide suitable context for the reader.
   - **Spatial**: Used only when precise locations or geographical patterns in data are more important to the reader than anything else.
   - **Ranking**: Use where an item's position in an ordered list is more important that its absolute or relative value. Don't be afraid to highlight the points of interest.
   - **Part-to-Whole**: Show how a single entity can be broken down into its component elements. If the reader's interest is solely in the size of the components, consider a magnitude-type chart instead.
   - **Flow**: Show the reader volumes or intensity of movement between two or more states or conditions. These might be logical sequences or geographical locations.
    

- **Relationship**: correlation, relationship, affect, change with, related
- **Comparison**: before/after, categories, compare, contrast, over time, peaks, rank, trend, types, valleys
- **Distribution**: cluster, distributed, from/to, plotted, points, spread, relative to, transfer
- **Composition**: components, group, makes up, of the whole, parts, percentage, pieces, portion, proportion, slices, total
- **Maps, Networks, Logic**: cluster, complex, connections, group, hierarchy, if/then, network, relationships, places, routes, structure, paths, organize, space, yes/no
