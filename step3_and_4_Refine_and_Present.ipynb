# Refine to Impress

### Structure & hierarchy
- Title: 12%
- Subtitle: 8%
- Visual field: 75%
- Sourceline: 5%
- Alignment: minimize # of lines
- Stay on the page

### Clarity
- No extraneous elements: does each element add meaning?
- All elements support the visual: Does each element highlight the idea and not the structure?
- Remove ambiguity: Does each element have a single, unambigous purpose?
- Use conventions & metaphors: Take advantage of ideas we "intuitively" understand.
- Clarify with annotations
- Movement distracts
- Consider presentation resolution
- Provide reference points
- Time contextualizes
- Communicate true scale
- Provide overview and summaries
- Accommodate color blindness
- Use learned symbols
- Use learned structures
- Consider data comfort and expertise

### Simplicity
- Show only what is needed: is each element necessary, unique and rendered as simply as possible?
- Use simplest appropriate visualizations
- One form of emphasis per element
- Minimal # of colors
- Limit eye travel: Are labels and legends in close proximity? Am I using minimal alignment lines
- Smooth animation
- Remove junk, increase data-to-ink ratio
- Limit colors
- Color has meaning
- Use gradients appropriately
- Large canvas
- Rarely use 3D
- Leave room to breathe
- Keep it lightweight
- Details add up

# Refine to Persuade

### Hone in on the main idea

- "I need to convince them that..."
- Provide goals

### Make it stand out
- emphasize the main idea by adding visualization information that calls attention to it.
- isolate the main idea by reducing the number of unique attribute for all other elements.
- "Chartjunk" (tufte's term) is not always bad, as sometimes they enhance ability for others to remember, e.g.
- Gradual reveal
- Guide attention
- Promote causality
- Provide sufficient dimensions
- Use common language

### Adjust what's around it

- manipulate elements to complement or contrast the main point, to make it pop
- remove reference points or plotted data that distracts or dilutes the main idea
- add reference points or plotted data to explose otherwise hidden context.
- shift reference points
- Maximize contrast
- Zoom to navigate

**The most efficient and effective transfer of data is always our primary goal. There may be others methods that aren't as efficient but the audience is going to understand it better, or others that aren't as effective but the time to develop costs too much. The effectiveness is not the only thing to consider.**

### Be Ethical
- Truncated y-axis: emphasizes or exaggerates change
- Double y-axis: creates crossovers, matching curves or gaps that don't actually mean anything
- The Map: size of reagion doesn't necessarily reflect the data encoded in it.
- Broken scales show drama where it doesn't exist.
- Decoration can be deceiving.
- Showing data on two different scales can make for an apples-to-oranges comparison.
- Showing a correlation can imply causation. 
- Ignoring population size makes rates impossible to compare.

### An Effective Storytelling Technique
##### The Gradual Reveal
- When designing, it’s important to remember that not everyone has the same level of expertise. Nor do people always know exactly what they are looking for. Often, when too much information is put in front of someone at once, it can be overwhelming. To prevent information overload it’s to guide people through information, while still allowing them to explore, through a technique called gradual reveal. .
- Gradual reveal works by designing the order of the information so that one grouping of information can be shown, or focused on, at a time. To do that effectively, start by deciding what information is most important and what order it needs to be presented in to make sense. Place the most important information at the top, or in the place where someone would begin, and layer the information from there.
- A gradual reveal can mean two different things. In one instance, only the initial piece of information would be visible. As items are selected, more visuals are added to the page. In another instance, all visuals would be visible at all times, but the reveal comes as items are selected and filters and smaller groupings are revealed.*

##### Considerations:
- Reveals generally are left to right, top to bottom
- Most important information should go at the top
- Reveal information as user expresses interest
- Consider using filtering and interactions
- Consider timed transitions
- Think about how different selections will take users down different paths
- Build understanding as user moves through design
